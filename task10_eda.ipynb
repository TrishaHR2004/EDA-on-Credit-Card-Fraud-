```python
# Create IQR Detection Function
def detect_outliers_iqr(series):
    Q1 = series.quantile(0.25)
    Q3 = series.quantile(0.75)
    IQR = Q3 - Q1
    
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    
    return (series < lower) | (series > upper)
```

```python
# Create Outlier Flag Column for Amount
df["amount_outlier"] = detect_outliers_iqr(df["Amount"])
```

```python
# Count How Many Outliers You Found
df["amount_outlier"].value_counts()
```

```python
# Interpretation "About 11% of transactions fall outside the normal spending range,
#indicating a significant volume of high-risk or unusual transactions that may require monitoring or automated fraud scoring."
```

```python
# Handle Outliers (CAPPING STRATEGY)

#We will NOT delete them — because in fraud analysis:

# Removing outliers = removing fraud signals 
```

```python
# Calculate IQR Limits
Q1 = df["Amount"].quantile(0.25)
Q3 = df["Amount"].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

lower, upper
```

```python
# Lower Bound = -101.75  → effectively no negative transactions
# Upper Bound = 184.51  → most “normal” spending is below ₹185 (or $185)
# Therefore “Any transaction above ~185 is statistically unusual and 
#should be treated as higher risk or monitored for fraud.”
```

```python
# Cap Amount Values
df["Amount_capped"] = np.where(df["Amount"] > upper, upper,
                               np.where(df["Amount"] < lower, lower, df["Amount"]))
```

```python
# Verify New Distribution
plt.figure(figsize=(6,4))
plt.boxplot(df["Amount_capped"], vert=False)
plt.title("Capped Transaction Amount Boxplot")
plt.xlabel("Amount")
plt.show()
```

```python
# “After capping, the transaction amount distribution became more stable while preserving fraud-related variation, 
#improving reliability for correlation and modeling.”
```

```python
# Create Correlation Matrix
corr_matrix = df.corr(numeric_only=True)
```

```python
# Find Top Correlated Pairs
top_corr = corr_matrix.unstack().sort_values(ascending=False)
top_corr[1:10]
```

```python
# This is showing self-correlation:

#Every column is perfectly correlated with itself (correlation = 1.0)

#So your code is currently listing:

#Feature vs SAME feature
#Which is not useful for analysis.
```

```python
# Filter Out Self-Correlations
corr_matrix = df.corr(numeric_only=True)

top_corr = (
    corr_matrix
    .where(~np.eye(corr_matrix.shape[0], dtype=bool))
    .unstack()
    .dropna()
    .sort_values(ascending=False)
)

top_corr.head(10)
```
